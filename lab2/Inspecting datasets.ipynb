## Python NLTK

The Natural Language Toolkit (NLTK) is a python library used for natural language processing (NLP).

* Tokenization (splitting each word in a text on f.ex whitespace or punctuation)
* Remove stopwords (discarding words that are too frequent to be informative)
* Stemming (representing each word using its word stem by dropping common suffixes)
* Lemmatization (representing each word using its word stem)

NLTK docs: https://www.nltk.org/


```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk import sent_tokenize, word_tokenize
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
```
